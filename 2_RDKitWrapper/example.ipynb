# 化合物(SMILES)のFingerprintと記述子を簡便に利用する為のライブラリ
### RDKitのインストールが必要
### RDkitWrapper.pyを実行スクリプトと同じフォルダに入れておくこと

```python
#化合物処理の為の関数・クラス群
from RDKitWrapper import draw_SMILES,Fingerprint,RDKitDescriptors
```

# SMILESを化合物として表示する関数 draw_SMILES


```python
"""
引数 smiles: ""Cc1ccccc1"" など
"""
draw_SMILES("Cc1ccccc1")
```

# Fingerprintを計算するクラスFingerpint

```python
#クラスを使用する為の初期化処理
fp=Fingerprint()
```

# calc関数: smilesを入力すると、fingerprintを自動計算

```python
"""
入力: SMILES (文字列)
出力: 0,1のリスト(512次元) 次元等を変えることも可能(RDKitWrapper.pyを参照)
"""
fp.calc("C")
```

```python
#おかしな化合物の場合はエラーが返ってくる
fp.calc("a")
```

## 一気に複数のSMILESを計算したい場合は、calc_list関数を使う

```python
"""
入力: SMILESを文字列のリスト ["C","CC","Cc1ccccc1",]
出力: pandasのdataframe
"""

fp.calc_list(["C","CC"])
```

# Descriptorの計算

```python
#クラスを使用する為の初期化処理
desc=RDKitDescriptors()
```

# calc関数: smilesを入力すると、記述子を自動計算

```python
"""
入力: smiles
出力: 記述子 (辞書形式)
"""
desc.calc("C")
```

## 一気に複数のSMILESを計算したい場合は、calc_list関数を使う

```python
"""
入力: smiles
出力: pandasのdataframe 
"""
desc.calc_list(["C","CC","CCC"])
```
